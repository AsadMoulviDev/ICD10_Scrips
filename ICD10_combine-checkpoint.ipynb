# Break Down ICD10CM Codes into 3 Levels 

```python
#Function to define Level2 titles

from itertools import chain
import pandas as pd

def level2(lines2):

    
    des22 = []
    cat = []
    e = []
    for i in range(0,len(lines2)):
        str2=lines2[i][0:9].strip()
        letter = str2[0]
        des2 = lines2[i][9:]
        b = str2.replace(letter,'')
        c=b.split('-')
        d=range(int(c[0]),int(c[1])+1)
        e.append(list(d))
        test = list(d)
        output = [str(int) for int in test]
        out2 = [output.zfill(2) for output in output]
        aces = [letter + out2 for out2 in out2]
        cat.append(aces)
        
        for x in range(len(e[i])):
            des22.append(des2)
    
    flatten_list = list(chain.from_iterable(cat))
    
    dict = {'cat': flatten_list, 'level2Title':des22} 
    df = pd.DataFrame(dict)
    return df
```

```python
#Function to define Level3 titles

def level3(lines):
    full_code = []
    des = []
    for i in range(0,len(lines)-1):
        line = lines[i]
        full_code_pre = line[0:5].strip()
        full_code.append(full_code_pre)
        des.append(line[5:])
    dict = {'cat': full_code, 'level3Title':des} 
    df = pd.DataFrame(dict)
    return df
```

```python
f2 = open("C:/Users/asadyousuf/Documents/Level2codes.txt", "r")
df = pd.read_csv('C:/Users/asadyousuf/Desktop/icd10codes.csv',encoding= 'unicode_escape')
f4 = open("C:/Users/asadyousuf/Documents/level3codes.txt", "r")
f3 = f2.read()
f5 = f4.read()
lines2 = f3.split('\n')
lines3 = f5.split('\n')

#Level2 Table
df2 = level2(lines2)

#Level3 Table
df3 = level3(lines3)
```

```python
inner_merged_total = pd.merge(df, df2,on=["cat"])
final = pd.merge(inner_merged_total,df3, on = ['cat'])
final
```
