# Break Down ICDPCS Codes into 2 Levels

```python
import pandas as pd

def create_df(lines):
    full_code = []
    des = []
    code_1 =[]
    code_2 =[]
    for i in range(0,len(lines)-1):
        line = lines[i]
        full_code_pre = line[0:6].strip()
        full_code.append(full_code_pre)
        code_1.append(full_code_pre[0])
        code_2.append(full_code_pre[0:2])
        des.append(line[8:])
    dict = {'fullcode': full_code, 'des':des,'code1':code_1,'code2':code_2} 
    df = pd.DataFrame(dict)
    return df

def create_df_l1(lines):
    full_code = []
    des = []
    for i in range(0,len(lines)-1):
        line = lines[i]
        full_code_pre = line[0:2].strip()
        full_code.append(full_code_pre)
        des.append(line[2:])
    dict = {'code1': full_code, 'level1des':des} 
    df = pd.DataFrame(dict)
    return df
def create_df_l2(lines):
    full_code = []
    des = []
    for i in range(0,len(lines)-1):
        line = lines[i]
        full_code_pre = line[0:3].strip()
        full_code.append(full_code_pre)
        des.append(line[3:])
    dict = {'code2': full_code, 'level2des':des} 
    df = pd.DataFrame(dict)
    return df
```

```python
f = open("C:/Users/asadyousuf/Downloads/2022-ICD-10-Codes-File/icd10pcs_codes_2022.txt", "r")
f2 = f.read()
lines = f2.split('\n')
df = create_df(lines)

f3 = open("C:/Users/asadyousuf/Documents/level1pcs.txt", "r")
f4 = f3.read()
lines1 = f4.split('\n')
df2 = create_df_l1(lines1)

f5 = open("C:/Users/asadyousuf/Documents/Level2pcs.txt", "r")
f6 = f5.read()
lines2 = f6.split('\n')
df3 = create_df_l2(lines2)
```

```python
inner_merged_total = pd.merge(df, df2,on=["code1"])
final = pd.merge(inner_merged_total,df3, on = ['code2'])
final.to_csv('final_ICD10PCS_codes.csv')
final
```
